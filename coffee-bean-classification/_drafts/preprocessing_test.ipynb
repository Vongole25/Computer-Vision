```python
import os
from utils import img_wrapper  # 필요한 경우 경로를 조정하세요.

# ImgWrapper 클래스 인스턴스 생성
wrapper = img_wrapper.ImgWrapper(height=224, width=224)

input_directory = './bean_img/raw_data/defect_bean/defected_beans'
output_directory = './bean_img/defect_bean'

# 출력 디렉터리가 없으면 생성
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# 파일 처리 및 파일명 증가를 위한 카운터
file_counter = 1

for filename in os.listdir(input_directory):
    if filename.lower().endswith(('.jpg', '.jpeg', '.png')):
        img_path = os.path.join(input_directory, filename)

        # 이미지 불러오기
        img = wrapper.load_image(img_path)

        # 리사이즈 및 중앙 크롭 수행
        processed_img = wrapper(img)

        # 원본 저장
        processed_img_path = os.path.join(output_directory, f'{file_counter}.png')
        processed_img.save(processed_img_path)
        print(f'Processed and saved: {processed_img_path}')
        file_counter += 1

        # 좌우 반전 저장
        flipped_horizontal = wrapper.flip_image(processed_img, mode='horizontal')
        flipped_horizontal_path = os.path.join(output_directory, f'{file_counter}.png')
        flipped_horizontal.save(flipped_horizontal_path)
        print(f'Processed and saved: {flipped_horizontal_path}')
        file_counter += 1

        # 상하 반전 저장
        flipped_vertical = wrapper.flip_image(processed_img, mode='vertical')
        flipped_vertical_path = os.path.join(output_directory, f'{file_counter}.png')
        flipped_vertical.save(flipped_vertical_path)
        print(f'Processed and saved: {flipped_vertical_path}')
        file_counter += 1

        # 상하 좌우 반전 저장
        flipped_both = wrapper.flip_image(processed_img, mode='both')
        flipped_both_path = os.path.join(output_directory, f'{file_counter}.png')
        flipped_both.save(flipped_both_path)
        print(f'Processed and saved: {flipped_both_path}')
        file_counter += 1

    else:
      print("fail")
```
